```python
import pandas as pd

# Load the BTC.csv file
df = pd.read_csv('BTC.csv')
df2 = pd.read_csv('MSCI.csv')

# Display the first few rows of the DataFrame
print(df.head())
print(df2.head())

```

Dates should match each other

```python
#import pandas as pd
#Matching dates
# Assuming `df1` and `df2` are your DataFrames

# Convert the 'date' column in the first DataFrame
df['date'] = pd.to_datetime(df['date'])

# Convert the 'Date' column in the second DataFrame
df2['Date'] = pd.to_datetime(df2['Date'])

# Rename the 'Date' column in df2 to match df1 if needed
df2.rename(columns={'Date': 'date'}, inplace=True)

print(df['date'].head())
print(df2['date'].head())
```

```python


# Assuming your DataFrame is named `df` and already loaded
# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Filter rows where the year in the 'date' column is 2024
subset_2024 = df[df['date'].dt.year == 2024]

# Display the filtered DataFrame
print(subset_2024.head())
```

```python
import matplotlib.pyplot as plt

# Assuming `df` is already loaded and the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Plot the 'close' variable against the 'date'
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['close'], label='Close Price', linewidth=1)

# Add labels and title
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price', fontsize=12)
plt.title('BTC Close Prices Over Time', fontsize=14)
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
merged_df = pd.merge(df, df2, on='date', how='inner')  # Use 'inner', 'left', or 'right' as needed

# Convert 'Open' and 'open' columns to numeric format
merged_df['Open'] = merged_df['Open'].str.replace(',', '', regex=True)
merged_df['Open'] = pd.to_numeric(merged_df['Open'], errors='coerce')
merged_df['open'] = pd.to_numeric(merged_df['open'], errors='coerce')

# Check the datatypes of both columns
print(merged_df[['Open', 'open']].dtypes)

# Drop rows where either column has NaN (optional)
merged_df = merged_df.dropna(subset=['Open', 'open'])

# Display the first few rows to confirm
print(merged_df[['date', 'Open', 'open']].head())

# Check the datatypes of both columns
print(merged_df[['Open', 'open']].dtypes)
# Display the merged DataFrame
print(merged_df.head())



# Filter rows where the date is before 2022
filtered_df = merged_df[merged_df['date'] > '2022-01-01']

# Display the filtered DataFrame
print(filtered_df.head())


```

```python
import statsmodels.api as sm
# Define the dependent and independent variables
X = filtered_df['Open']  # Independent variable from the second dataset
y = filtered_df['open']  # Dependent variable from the first dataset

# Add a constant to the independent variable (required by statsmodels)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the linear model
print(model.summary())
```
